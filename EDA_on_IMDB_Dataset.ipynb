#Convert the string values to proper lists and apply the transformation to 3 columns namely (genres, crew,cast)

```python
def ctl(st):   #eval changes the string to list in this case
  return eval(st)
```

```python
#apply the above function on three columns and choose condition where we only apply these to the non null values
imdb_df.loc[~imdb_df['genres'].isna(),'genres']=imdb_df.loc[~imdb_df['genres'].isna(),'genres'].apply(ctl)
```

```python
imdb_df.loc[0,'genres']
```

```python
imdb_df.loc[~imdb_df['cast'].isna(),'cast']=imdb_df.loc[~imdb_df['cast'].isna(),'cast'].apply(ctl)  #same for the cast column
```

```python
imdb_df.loc[~imdb_df['crew'].isna(),'crew']=imdb_df.loc[~imdb_df['crew'].isna(),'crew'].apply(ctl)#same for the crew column
```

## 3. Finding out the unique genres of movies in the dataset

```python
#Considering only those rows in which genres column is not null
imdb_df_no_na_genres=imdb_df_new[~imdb_df_new['genres'].isna()]
```

```python
imdb_df_no_na_genres.shape
```

```python
imdb_df_no_na_genres.loc[0,'genres']
```

```python
#Create the list of genres
genres_list=[]

# Iterate over each observation in a dataframe
#The iterrows function works like dict.vlaues in dictionary. The index takes index values of dataframe and the row takes  all the other columns present in it

for index, row in imdb_df_no_na_genres.iterrows(): 

  # Identify the list of genres
  genre_list=imdb_df_no_na_genres.loc[index, 'genres']  

  # We will iterate over list of dictionaries
  for elem in genre_list:
    genres_list.append(elem['name'])
```

```python
set(genres_list)     #set gives unique set of genres
```
